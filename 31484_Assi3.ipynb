```python
meanNumber = data["Number"].mean()
data["Number"] = data["Number"].fillna(meanNumber)
```

```python
meanNumber = data["Age"].mean()
data["Age"] = data["Age"].fillna(meanNumber)
```

```python
meanNumber = data["Weight"].mean()
data["Weight"] = data["Weight"].fillna(meanNumber)
```

```python
meanNumber = data["Salary"].mean()
data["Salary"] = data["Salary"].fillna(meanNumber)
```

```python
mode = data["Name"].mode()[0]
data["Name"] = data["Name"].fillna(mode)
```

```python
mode = data["Team"].mode()[0]
data["Team"] = data["Team"].fillna(mode)
```

```python
mode = data["Position"].mode()[0]
data["Position"] = data["Position"].fillna(mode)
```

```python
mode = data["Height"].mode()[0]
data["Height"] = data["Height"].fillna(mode)
```

```python
mode = data["College"].mode()[0]
data["College"] = data["College"].fillna(mode)
```

```python
data.isnull().sum()
```

```python
import numpy as np
std = np.std(data.Age)
std
```

```python
data.Age.mode()[0]
```

```python
data["Age"].value_counts()
```

```python
bins = [14, 19, 24, 29, 34, 39, 44]
labels = ['14-18', '19-23', '24-28', '29-33', '34-38', '39-43']
data["Age Groups"] = pd.cut(data.Age, bins = bins, labels = labels)
data["Age Groups"].value_counts()
```

```python
data["Age Groups"].describe()
```

```python
ageGrp = data.groupby("Age Groups")
ageGrp.get_group("24-28").Salary.describe()
```

```python
ageGrp["Salary"].describe()
```
