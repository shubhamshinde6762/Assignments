```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Iris.csv")
df
```

```python
df.dtypes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,10))
df.drop('Id', axis=1).hist(bins=20, figsize=(15,10))
plt.tight_layout()
plt.show()

plt.figure(figsize=(15,10))
df.drop(['Id', 'Species'], axis=1).boxplot()
plt.title('Box Plot of Numeric Features')
plt.show()
```

```python
df.Species.value_counts()
```

```python
plt.figure(figsize=(8,8))
df['Species'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Species Proportion in Iris Dataset')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,10))
numeric_features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

for i, feature in enumerate(numeric_features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='Species', y=feature, data=df)
    plt.title(f'{feature} Distribution by Species')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

plt.figure(figsize=(15,10))
for i, feature in enumerate(numeric_features, 1):
    plt.subplot(2, 2, i)
    sns.violinplot(x='Species', y=feature, data=df)
    plt.title(f'{feature} Violin Plot by Species')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
