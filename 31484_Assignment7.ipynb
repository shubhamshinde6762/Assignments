```python
import spacy
from sklearn.feature_extraction.text import TfidfVectorizer
from docx import Document
import string

nlp = spacy.load("en_core_web_sm")

doc = Document('testdoc.docx')
text = ' '.join(p.text for p in doc.paragraphs)
print(text)

docSpacy = nlp(text)
print(docSpacy)

tokens = [token.text.lower() for token in docSpacy if not token.is_stop and not token.is_punct]
print("tokens \n", tokens)
pos_tags = [(token.text, token.pos_) for token in docSpacy]
lemmas = [token.lemma_ for token in docSpacy if not token.is_stop and not token.is_punct]

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform([' '.join(tokens)])
terms = tfidf.get_feature_names_out()
scores = dict(zip(terms, tfidf_matrix.toarray()[0]))

print(f"\nTotal sentences: {len(list(docSpacy.sents))}")
print(f"Total words: {len(tokens)}")
print(f"\nPOS tags:", pos_tags)
print(f"\nlemmas:", lemmas)
print("\nterms by TF-IDF:")
for term, score in sorted(scores.items(), key=lambda x: x[1], reverse=True):
    print(f"{term}: {score:.8f}")
```
