```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("StudentsPerformance_modified.csv")
data
```

```python
data.isnull().sum().sort_values(ascending = False)
```

```python
data = data.dropna()
data
```

```python
data.isnull().sum().sort_values(ascending = False)
```

```python
data.dtypes
```

```python
# data = data.fillna(data.mean(), inplace=True)
data['math score'] = pd.to_numeric(data['math score'], errors='coerce').dropna(0).astype(np.float64)
data.dtypes
```

```python
data = data[(data['math score'] >= 0) & (data['reading score'] >= 0) & (data['writing score'] >= 0)]
```

```python
import scipy.stats as stats
data[['zmath', 'zreading', 'zwriting']] = data[['math score', 'reading score', 'writing score']].apply(stats.zscore)
```

```python
data
```

```python
outliers = data[(abs(data['zmath']) > 3) | (abs(data['zreading']) > 3) | (abs(data['zwriting']) > 3)]
data = data[(abs(data['zmath']) <= 3) & abs(data['zreading']) <= 3 & (abs(data['zwriting']) <= 3)]
outliers
```

```python
data
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data[['math score', 'reading score', 'writing score']] = scaler.fit_transform(data[['math score', 'reading score', 'writing score']])
data
```
